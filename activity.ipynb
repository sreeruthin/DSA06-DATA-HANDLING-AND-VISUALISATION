```r
data <- c(40, 20, 30, 10)
labels <- c("A", "B", "C", "D")
pie(data, labels = labels, main = "Pie Chart Example")
```

```r
data <- data.frame(
  category = c("A", "B", "C"),
  group1 = c(10, 20, 30),
  group2 = c(15, 25, 35)
)
library(ggplot2)
ggplot(data, aes(x = category)) +
  geom_bar(aes(y = group1), stat = "identity", position = "dodge", fill = "blue") +
  geom_bar(aes(y = group2), stat = "identity", position = "dodge", fill = "red") +
  labs(title = "Side-by-Side Bar Chart")
```

```r
data <- data.frame(
  category = c("A", "B", "C"),
  subcategory = c("A1", "B1", "C1"),
  value = c(10, 20, 30)
)
library(ggplot2)
ggplot(data, aes(x = "", y = value, fill = subcategory)) +
  geom_bar(stat = "identity", width = 1) +
  coord_polar(theta = "y") +
  labs(title = "Nested Pie Chart")
```

```r
data <- data.frame(
  x = c(1, 2, 3, 4, 5),
  y = c(2, 4, 5, 7, 8)
)

ggplot(data, aes(x = x, y = y)) +
  geom_point() +
  labs(title = "Scatterplot Example")
```
